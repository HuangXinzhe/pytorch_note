# 深度学习必要部分
- 机器学习/深度学习任务的整体流程
- 在机器学习/深度学习的每个部分的作用

- 数据的预处理
    - 数据格式统一
    - 异常数据的消除
    - 必要的数据变换
    - 划分训练集、验证集和测试集
- 模型的选择
    - 损失函数
    - 优化方法
    - 超参数
- 模型训练

深度学习所需的样本量很大，一次加载全部数据运行可能会超出内存容量而无法实现；同时还有批（batch）训练等提高模型表现的策略，需要每次训练读取固定数量的样本送入模型中训练

深度学习中训练和验证过程最大的特点在于读入数据是按批的，每次读入一个批次的数据，放入GPU中训练，然后将损失函数反向传播回网络最前面的层，同时使用优化器调整网络参数。这里会涉及到各个模块配合的问题。训练/验证后还需要根据设定好的指标计算模型表现。
